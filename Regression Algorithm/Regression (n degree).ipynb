```python
import random
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt

class nDegPred():
    def __init__(self, m, degree):
        self.m = m
        self.degree = degree
    def calculate(self):        
        self.a = np.random.normal(0,50,self.m)
        
        self.a2 = []
        self.b = np.random.normal(0,50,self.m)
        A = np.array([[1]*self.m, self.a])
        
        for i in range(2, self.degree + 1):
            for n in self.a:
                self.a2.append(n**i)
            A = np.vstack([A, self.a2])
            self.a2.clear()

        A = A.T
        #A.T*A x^ = A^Tb
        print(A.shape)
        x = (inv(A.T.dot(A)).dot(A.T)).dot(self.b)

        # y = C + Dt
        

        self.z = np.linspace(min(self.a), max(self.a), 100)
        
        self.y = 0
        for i in range(self.degree):
            self.y += x[self.degree] * (self.z**(self.degree))
        
        #self.y = E*((self.z)**2)  + D*self.z + C
    def affichage(self):
        plt.scatter(self.a, self.b)
        plt.plot(self.z, self.y, '-r', label='regression')
        plt.title(f"Degree: {self.degree}")
        plt.ylim([-100, 100])
        plt.xlim([-100, 100])
```

```python
pred = nDegPred(4, 4)
pred.calculate()
pred.affichage()
```
