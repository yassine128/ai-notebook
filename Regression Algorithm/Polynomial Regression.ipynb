```python
import random
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt

class PolyReg():
    def __init__(self, m):
        self.m = m
    def calculate(self):        
        self.a = np.random.normal(0,50,self.m)
        
        self.a2 = []
        
        for n in self.a:
            self.a2.append(n**2)
        
        self.b = np.random.normal(0,50,self.m)

        A = np.array([[1]*self.m, self.a, self.a2]).T
        #A.T*A x^ = A^Tb

        x = (inv(A.T.dot(A)).dot(A.T)).dot(self.b)

        # y = C + Dt
        
        E = x[2]
        C = x[0]
        D = x[1]

        self.z = np.linspace(min(self.a), max(self.a), 100)
        self.y = E*((self.z)**2)  + D*self.z + C
    def affichage(self):
        plt.scatter(self.a, self.b)
        plt.plot(self.z, self.y, '-r', label='regression')
        plt.ylim([-70, 70])
```

```python
pred = PolyReg(4)
pred.calculate()
pred.affichage()
```
